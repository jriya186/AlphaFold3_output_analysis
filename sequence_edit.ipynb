### Glycosylation site determination

```python
# enter desired glycosylation site/s, protein start and protein end)
def map_glycosylation_sites(full_length_sites, trunc_start, trunc_end):
    mapped = []
    for site in full_length_sites:
        if trunc_start <= site <= trunc_end:
            mapped_pos = site - trunc_start + 1
            mapped.append((site, mapped_pos))
        else:
            print(f"Site {site} is outside the truncated region ({trunc_start}-{trunc_end})")
    return mapped
```
